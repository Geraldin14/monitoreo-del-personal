```python
# Análisis estadístico

import pandas as pd

# Replace 'your_excel_file.xlsx' with the actual name of your Excel file
excel_file = pd.ExcelFile('data.xlsx')

# Cargar la hoja "Data"
df_data = excel_file.parse("Data")

# Mostrar las primeras filas para explorar los datos
display(df_data.head())

# Número de registros
num_registros = df_data.shape[0]

# Distribución por sexo, área y sede
sexo_counts = df_data['Sexo'].value_counts()
area_counts = df_data['Area'].value_counts()
sede_counts = df_data['Sede'].value_counts()

# Estadísticas del salario
salario_stats = df_data['Salario'].describe()

# Estadísticas de satisfacción y objetivos
satisfaccion_counts = df_data['Satisfacción'].value_counts()
objetivo_counts = df_data['Objetivo'].value_counts()

# Frecuencia de los indicadores (1 al 4 como ejemplo)
indicadores_freq = {}
for i in range(1, 5):
    col = f'Indicador {i}'
    if col in df_data.columns:
        indicadores_freq[col] = df_data[col].value_counts()

results = {
    "Total de registros": num_registros,
    "Distribución por sexo": sexo_counts.to_dict(),
    "Distribución por área": area_counts.to_dict(),
    "Distribución por sede": sede_counts.to_dict(),
    "Estadísticas salariales": salario_stats.to_dict(),
    "Satisfacción": satisfaccion_counts.to_dict(),
    "Objetivo": objetivo_counts.to_dict(),
    "Frecuencia de Indicadores": {k: v.to_dict() for k, v in indicadores_freq.items()}
}

display(results)
```
